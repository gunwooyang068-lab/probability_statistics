```python
import numpy as np
#numpy << 수학 계산, 배열을 다룰 때 쓰는 라이브러리임
#np << 수학 계산용
import matplotlib.pyplot as plt
#metapltlib.pyplot << 그래프 그릴 때 쓰는 라이브러리임
#plt << 그래프 그리기용

x=np.array([1,2,3,4,5,6,7])
y=np.array([78.4,78.6,78.5,78.9,78.2,79.4,78.6])

plt.plot(x,y,color="r")
#plt.plot() << 선그래프를 그리는 함수
#color="r" << 선의 색을 빨간색으로 지정함 ('r'ed)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5,6,7])
y=np.array([78.4,78.6,78.5,78.9,78.2,79.4,78.6])

plt.plot(x,y,color="r",marker='o')
plt.grid()
#plt.grid() << 그래프에 격자를 표시해줌
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace (-5,5,300)
#-5부터 5까지를 300개의 점을오 쪼개서 배열을 생성
#곡선을 부드럽게 그리기 위해 데이터 포인트를 넣는 역할임
y1=x**2
#원점을 꼭짓점으로 하는 포물선
y2=(x-2)**2
#오른쪽으로 2만큼 평행이동한 포물선

plt.plot(x,y1,color="r")
plt.plot(x,y2,color="k",linestyle="--")
#-- << 점선

plt.grid()
plt.show()
```
